```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/train.csv')
df.shape
```

```python
# 删除包含'gyro'的列
gyro_cols = [col for col in df.columns if 'Gyro' in col]
df.drop(columns=gyro_cols, inplace=True)
df.shape
```

```python
df.to_csv('data/train_no_gyro.csv', index=False)
```

```python
data = df

# 提取所需的列
columns_of_interest = ['tBodyAcc-mean()-X', 'tBodyAcc-mean()-Y', 'tBodyAcc-mean()-Z']
extracted_data = data[columns_of_interest]

# 绘制线性图表
plt.figure(figsize=(12, 6))
plt.plot(extracted_data['tBodyAcc-mean()-X'], label='tBodyAcc-mean()-X')
plt.plot(extracted_data['tBodyAcc-mean()-Y'], label='tBodyAcc-mean()-Y')
plt.plot(extracted_data['tBodyAcc-mean()-Z'], label='tBodyAcc-mean()-Z')

# 添加图例
plt.legend()

# 添加标题和标签
plt.title('Time Series of tBodyAcc-mean()-X, Y, Z')
plt.xlabel('Sample Index')
plt.ylabel('Acceleration')

# 显示图表
plt.show()
```
