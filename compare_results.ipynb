```python
import seaborn as sns
import pandas as pd

# 两个列表数据作为示例
models = ['SVM', 'DNN', 'CNN', 'SVM', 'DNN', 'CNN']
state = ['Before', 'Before', 'Before', 'After', 'After', 'After']
accuracy = [96, 96, 99, 91, 92, 94]

# 将列表数据转换为DataFrame
data = pd.DataFrame({'models': models, 'state': state, 'accuracy': accuracy})

# 设置Seaborn主题
sns.set_theme(style="whitegrid")

# 画出嵌套的柱状图
g = sns.catplot(
    data=data, kind="bar",
    x="models", y="accuracy", hue="state",
    ci="sd", palette="dark", alpha=.6, height=6
)
g.set_axis_labels("Model", "Accuracy (%)")
g.despine(left=True)
g.set_axis_labels("", "Accuracy (%)")
g.legend.set_title("STATE")
```
